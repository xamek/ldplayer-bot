```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
from typing import List

def show_image(img, title="Image"):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()

# Load screenshot
screenshot = cv2.imread(
    r"D:\Xamek\Workspace\ldplayer-bot\unknown_states\unknown_20251226_085233_147.png"
)
show_image(screenshot, "Screenshot")

def test_template_modes(screenshot, template, threshold=0.7):
    """Compare template matching in color, grayscale, and edge detection."""

    # --- Color mode ---
    result_color = cv2.matchTemplate(screenshot, template, cv2.TM_CCOEFF_NORMED)
    _, max_val_color, _, max_loc_color = cv2.minMaxLoc(result_color)
    print(f"[COLOR] Confidence: {max_val_color:.3f}")

    # --- Grayscale mode ---
    screenshot_gray = cv2.cvtColor(screenshot, cv2.COLOR_BGR2GRAY)
    template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
    result_gray = cv2.matchTemplate(screenshot_gray, template_gray, cv2.TM_CCOEFF_NORMED)
    _, max_val_gray, _, max_loc_gray = cv2.minMaxLoc(result_gray)
    print(f"[GRAY]  Confidence: {max_val_gray:.3f}")

    # --- Edge detection mode ---
    screenshot_edges = cv2.Canny(screenshot_gray, 100, 200)
    template_edges = cv2.Canny(template_gray, 100, 200)
    result_edges = cv2.matchTemplate(screenshot_edges, template_edges, cv2.TM_CCOEFF_NORMED)
    _, max_val_edges, _, max_loc_edges = cv2.minMaxLoc(result_edges)
    print(f"[EDGES] Confidence: {max_val_edges:.3f}")

    # --- Pick best mode ---
    scores = {
        "COLOR": (max_val_color, max_loc_color, template.shape[:2]),
        "GRAY": (max_val_gray, max_loc_gray, template_gray.shape),
        "EDGES": (max_val_edges, max_loc_edges, template_edges.shape),
    }
    best_mode, (best_val, best_loc, best_size) = max(scores.items(), key=lambda x: x[1][0])

    if best_val >= threshold:
        h, w = best_size
        top_left = best_loc
        bottom_right = (top_left[0] + w, top_left[1] + h)
        screenshot_copy = screenshot.copy()
        cv2.rectangle(screenshot_copy, top_left, bottom_right, (0,255,0), 2)
        show_image(screenshot_copy, f"Best match ({best_mode}): {best_val:.3f}")
    else:
        print("No match above threshold")

def get_templates_from_dir(directory_path: str) -> List[str]:
    """Get all .png files from a directory as absolute paths."""
    patterns = []
    if os.path.exists(directory_path):
        for filename in os.listdir(directory_path):
            if filename.endswith('.png'):
                patterns.append(os.path.join(directory_path, filename))
                print(f"filename: {filename}")
    return patterns

# Path to your templates folder
LOADING_TEMPLATES_DIR = os.path.join(
    r"D:\Xamek\Workspace\ldplayer-bot\states\main_menu", "templates"
)

loading_patterns = get_templates_from_dir(LOADING_TEMPLATES_DIR)

# Batch test all templates with color, grayscale, and edge comparison
for t in loading_patterns:
    template = cv2.imread(t)
    show_image(template, "Template")
    print(f"Testing {t}...")
    test_template_modes(screenshot.copy(), template, threshold=0.7)
```
